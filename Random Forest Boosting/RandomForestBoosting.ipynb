```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
from sklearn.metrics import precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
from sklearn import metrics

import seaborn as sns
```

```python
X = train.drop('Class', axis=1)
y = train['Class']
```

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.2, random_state=24)
print(train_X.shape, test_X.shape, train_y.shape, test_y.shape)
```

```python
def Train_Fit_Report(model, train_X, train_y, test_X, test_y):
    model.fit(train_X, train_y)
    train_pred = model.predict(train_X)
    test_pred = model.predict(test_X)
    
    print("Train Accuracy:", accuracy_score(train_y, train_pred))
    print("Test Accuracy:", accuracy_score(test_y, test_pred))
    
    print("Train Classification Report:\n", classification_report(train_y, train_pred))
    print("Test Classification Report:\n", classification_report(test_y, test_pred))
    
    print("Train Confusion Matrix:\n", confusion_matrix(train_y, train_pred))
    print("Test Confusion Matrix:\n", confusion_matrix(test_y, test_pred))
    return model
    
```

```python
#Home work.1 to do a GRID SEARCH CV for hyperparameter tuning
decisionTree_imbalanced = DecisionTreeClassifier()
Train_Fit_Report(decisionTree_imbalanced, train_X, train_y, test_X, test_y)
```

```python
decisionTree_roughTuned = DecisionTreeClassifier(max_depth=5)
Train_Fit_Report(decisionTree_roughTuned, train_X, train_y, test_X, test_y)
```

## Random Forest Parameters
1. n_estimators --> How many trees you want to build
2. max_depth --> How deep each tree should be
3. min_samples_split --> Minimum samples required to split a node
4. min_samples_leaf --> Minimum samples required at each leaf node
5. max_features --> Number of features to consider at each split
6. criterion --> 'gini' or 'entropy' for classification
7. Class_weight --> 'balanced' to handle class imbalance, which mistakes in classification are more costly 
8. Bootstrap --> Whether to use bootstrap samples or not

```python
# Home work.2 to add parameters for Random Forest Classifier
randomForest = RandomForestClassifier()
randomForest = Train_Fit_Report(randomForest, train_X, train_y, test_X, test_y)
```

## Ada Boost Parameters:
1. n_estimators: How many corrective rounds should model perform
2. learning_rate: How strongly should each tree correct the previous mistakes / How much influence should each tree have when correct past mistakes.
3. base_estimator: 

```python
adaBoost = AdaBoostClassifier(n_estimators=100, learning_rate=0.5)
adaBoost = Train_Fit_Report(adaBoost, train_X, train_y, test_X, test_y)
```

## Observations of AdaBoost:
1. High Learning Rate = Overfitting
2. Too Many estimators also include risk of overfitting

```python
gbmBoost = GradientBoostingClassifier(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=24)
gbmBoost = Train_Fit_Report(gbmBoost, train_X, train_y, test_X, test_y)
```

Observations of Gradient Boost:
1. Slower than Randomforest
2. Careful tuning is very very important

## Parameters for XG Boost:
1. n_estimators - How many trees?
2. learning_rate - How fast do we learn?
3. max_depth - how complex is each tree?
4. subsample - How much data does each tree see?
5. colsample_bytree - How many features per tree?
6. reg_alpha & reg_lamda - How much do we punish complexity
7. scale_pos_weight - How much more important is fraud?
8. min_child_weight - How much data must a branch have to be not pruned?
9. gamma - IS this current split worth it? - forces you to have meaningful improvements only - Minimum benifit requrement.
10. colsample_bylevel - how man feature per tree level?

```python
xgbBoost = XGBClassifier()
xgbBoost = Train_Fit_Report(xgbBoost, train_X, train_y, test_X, test_y)
```

## Observations of XG Extensive Gradient Boost:
1. Out of all ensemble models XGB performes best
2. XGB is prod version of GBM
3. Regularization (l1/l2)
4. Extreme pruing can be done via Gamma


```python
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import RandomOverSampler
scaler = StandardScaler()

X_scaled = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
train_X_scaled, test_X_scaled, train_y_scaled, test_y_scaled = train_test_split(X_scaled, y, test_size=0.2, random_state=24)
```

```python
randomSampler = RandomOverSampler(random_state=24)
X_resampled, y_resampled = randomSampler.fit_resample(train_X_scaled, train_y_scaled)

print(train_y_scaled.value_counts())
print(y_resampled.value_counts())
```

```python
print("After Scaling and Resampling")
print("-----------------------------------------------------------------")
print("Decision Tree Classifier Results")
print("-----------------------------------------------------------------")
Train_Fit_Report(decisionTree_imbalanced, X_resampled, y_resampled, test_X_scaled, test_y_scaled)
print("-----------------------------------------------------------------")
print("Decision Tree Classifier Rougn Tuned Results")
print("-----------------------------------------------------------------")
Train_Fit_Report(decisionTree_roughTuned, X_resampled, y_resampled, test_X_scaled, test_y_scaled)
print("-----------------------------------------------------------------")
print("Random Forest Classifier Results")
print("-----------------------------------------------------------------")
Train_Fit_Report(randomForest, X_resampled, y_resampled, test_X_scaled, test_y_scaled)
print("-----------------------------------------------------------------")
print("Ada Boost Classifier Results")
print("-----------------------------------------------------------------")
Train_Fit_Report(adaBoost, X_resampled, y_resampled, test_X_scaled, test_y_scaled)
print("-----------------------------------------------------------------")
print("Gradient Boosting Classifier Results")
print("-----------------------------------------------------------------")
Train_Fit_Report(gbmBoost, X_resampled, y_resampled, test_X_scaled, test_y_scaled)
print("-----------------------------------------------------------------")
print("XG Boost Classifier Results")
print("-----------------------------------------------------------------")
Train_Fit_Report(xgbBoost, X_resampled, y_resampled, test_X_scaled, test_y_scaled)
```
